Для ряда со средним дневным количеством молока на корову из предыдущего вопроса давайте с помощью критерия Дики-Фуллера подберём порядок дифференцирования, при котором ряд становится стационарным.

**Дифференцирование** (переход к попарным разностям соседних значений) можно делать так:

Ряд можно будет считать Стационарным если критерий ДФ<0.05

Чтобы сделать **сезонное дифференцирование**, нужно изменить значение параметра у функции shift:

**Примечание:** советуют сначала сделать сезонное преобразование (оно может помочь сразу)

```python
df['daily_diff_only12'] = df.avg_day - df.avg_day.shift(12)
# нужно удалить значения Nan иначе ряд сходиться не будет(!)
df_only12 = df.dropna(subset=['daily_diff_only12'])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(df_only12.daily_diff_only12)[1])
```

```python
plt.figure(figsize(15,7))
df_only12.daily_diff_only12.plot()
plt.ylabel('Avg milk 1')
pylab.show()
```

```python
df['daily_diff1'] = df_only12.daily_diff_only12 -  df_only12.daily_diff_only12.shift(1)
df1 = df.dropna(subset=['daily_diff1'])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(df1.daily_diff1)[1])
```

```python
plt.figure(figsize(15,7))
df1.daily_diff1.plot()
plt.ylabel('Avg milk 2')
pylab.show()
```

Для стационарного ряда из предыдущего вопроса (продифференцированного столько раз, сколько вы посчитали нужным) постройте **график автокорреляционной функции**. Это можно cделать так:

Исходя из этого графика, какое начальное приближение вы предложили бы для параметра *Q* в модели SARIMA? 0 (Описанные значения кратны 12. Обратите внимание, что значений, не кратных периоду сезонности, на графике отсутствуют. Это и есть ответ.)

Начальное значение для параметра Q ∗ S даёт номер последнего сезонного лага, при котором автокорреляция значима.

```python
sm.graphics.tsa.plot_acf(df1.daily_diff1.values.squeeze(), lags=50)
```

Для того же ряда, что и в предыдущем вопросе, постройте график **частичной автокорреляционной функции**. Это можно сделать так:

Исходя из этого графика, какое начальное приближение вы предложили бы для параметра *p* в модели SARIMA? 1 (12/12)

Начальное приближение для параметра P ∗S задаёт номер последнего сезонного лага, при котором частичная автокорреляция значима. В данных из примера это лаг под номером 24, значит начальное приближение
P = 2, поскольку длина сезонного периода S = 12. Аналогично, p задаётся как номер последнего несезонного
лага, при котором частичная автокорреляция значима.

```python
sm.graphics.tsa.plot_pacf(df1.daily_diff1.values.squeeze(), lags=50)
```
