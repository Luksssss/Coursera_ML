## sklearn.grid_search

документация: http://scikit-learn.org/stable/modules/grid_search.html

```python
from sklearn import model_selection, datasets, linear_model, metrics
# grid_search, 

import numpy as np
import pandas as pd
```

### Генерация датасета

```python
iris = datasets.load_iris()
```

```python
train_data, test_data, train_labels, test_labels = model_selection.train_test_split(iris.data, iris.target, 
                                                                                     test_size = 0.3,random_state = 0)
```

### Задание модели

```python
# стохастический градиентный спуск
classifier = linear_model.SGDClassifier(random_state = 0)
```

### Генерация сетки

```python
classifier.get_params().keys()
```

```python
# loss - вид функции потерь
parameters_grid = {
    'loss' : ['hinge', 'log', 'squared_hinge', 'squared_loss'],
    'penalty' : ['l1', 'l2'],
    'n_iter' : range(5,10),
    'alpha' : np.linspace(0.0001, 0.001, num = 5),
}
```

```python
cv = model_selection.StratifiedShuffleSplit(n_splits = 10, test_size = 0.2, random_state = 0)
# cv.get_n_splits(train_data, train_labels)

# cv.split(train_data, train_labels)
# cv = model_selection.StratifiedShuffleSplit(train_labels, n_splits = 10, test_size = 0.2, random_state = 0)
```

### Подбор параметров и оценка качества

#### Grid search

```python
# задаём сетку параметров
# scoring - метрика по которой будем искать наилучшее решение
grid_cv = model_selection.GridSearchCV(classifier, parameters_grid, scoring = 'accuracy', cv = cv)
```

```python
%%time
grid_cv.fit(train_data, train_labels)
```

```python
grid_cv.best_estimator_
```

```python
print grid_cv.best_score_
print grid_cv.best_params_
```

```python
# grid_cv.grid_scores_[:10]
# все возможные задействованные варианты значений сетки
grid_cv.cv_results_['params'][:10]

# среднее в каждом проходе
grid_cv.cv_results_['mean_train_score'][:10]

# стандартное отклонение каждого прохода
grid_cv.cv_results_['std_train_score'][:10]
```

#### Randomized grid search (случайный поиск по сетке)
выгодно использовать когда у нас модель долго обучается

```python
# n_iter - количество случайных элементов в сетке которые мы будем вычислять
randomized_grid_cv = model_selection.RandomizedSearchCV(classifier, parameters_grid, scoring = 'accuracy', cv = cv, n_iter = 20, 
                                                   random_state = 0)
```

```python
%%time
randomized_grid_cv.fit(train_data, train_labels)
```

```python
print randomized_grid_cv.best_score_
print randomized_grid_cv.best_params_
```
