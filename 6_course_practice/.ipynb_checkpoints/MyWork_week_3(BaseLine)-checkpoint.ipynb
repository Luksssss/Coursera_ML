```python
%pylab inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

from sklearn import model_selection, datasets, metrics
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.tree import export_graphviz
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('orange_small_churn_data.train', sep = ',', header = 0)
labels = pd.read_csv('orange_small_churn_labels.train', sep = ',', header = None, names = ['label'])
print data.shape, labels.shape
data.head()
```

## Подготовка данных

```python
# мешаем, выделяем тестовую и тренировочную выборку
train_data, test_data, train_labels, test_labels = model_selection.train_test_split(data, labels,
                                                                                     shuffle = True,
                                                                                     test_size = 0.2)
```

```python
# отделяем числовые и категориальные признаки и удаляем полностью пустые признаки
data_numb = train_data.iloc[:,0:190].dropna(axis=1, how='all')
data_categ = train_data.iloc[:,190:].dropna(axis=1, how='all')

# масштабируем числовые признаки (старый способ)
# data_numb = (data_numb-data_numb.min(axis=0))/(data_numb.max(axis=0)-data_numb.min(axis=0))

# NOTE: сделаем скелинг численных признаков с D=1 (подробнее см. неделю 1)
# NOTE2: такая запись по созданию нового DataFrame нужна из-за тоог что при fit_transform слетают индексы и потом
# при контатенации у нас получается каша.
scaler = StandardScaler()
data_numb=pd.DataFrame(scaler.fit_transform(data_numb.values), index=data_numb.index)
# scaled_matrix = scaler.fit_transform(data_numb.values)
# data_numb = pd.DataFrame(scaled_matrix, columns=data_numb.columns)


# удалим те колонки где больше 20 категорий и меньше 2 (эти колонки не информативны)
name_del = [name for name, var in data_categ.iteritems() if var.value_counts(dropna=True).shape[0] > 20 or var.value_counts(dropna=True).shape[0] < 2]
data_categ = data_categ.drop(labels=name_del, axis=1)
```

```python
# Замена на числовой индекс значения
# data_categ.fillna('', inplace=True)

# enc = OrdinalEncoder()
# enc.fit(data_categ)
# # enc.categories_
# dc = enc.transform(data_categ)

# кодируем категориальные признаки
data_dummies = pd.get_dummies(data_categ)
```

```python
print data_numb.shape
print data_dummies.shape
print train_labels.shape
```

```python
# объединяем числовые признаки и закодированные категориальные
train_data = pd.concat([data_numb, data_dummies], axis=1)
train_data.fillna(0, inplace=True)
```

```python
print train_data.shape
print train_labels.shape
print type(train_data)
```

## Линейная модель

```python
from sklearn.linear_model import RidgeClassifier

parameters_grid = {
    'alpha' : np.linspace(0.1, 1, num = 10)
}

classifier = RidgeClassifier(random_state=0)

# Будем использовать метод стратификации который делит соотношение классов в обучающей выборке на равное количество
skf = model_selection.StratifiedKFold(n_splits = 5, shuffle = True, random_state = 0)

grid_cv = model_selection.GridSearchCV(classifier, parameters_grid, scoring = 'roc_auc', cv=skf)
```

```python
grid_cv.fit(train_data, train_labels)
```

```python
grid_cv.best_estimator_
```

```python
print 'ROC_AUC = %f' % grid_cv.best_score_
print grid_cv.best_params_
# print grid_cv.cv_results_
```

```python
clf = RidgeClassifier(random_state=0, alpha=1)
clf.fit(train_data, train_labels)
actual_labels = clf.predict(train_data)
```

```python
print "accuracy =", clf.score(train_data, train_labels)
print "precision =", metrics.precision_score(train_labels, actual_labels)
print "recall =", metrics.recall_score(train_labels, actual_labels)
```

```python
(actual_labels == 1).sum()
```

**NOTE:** Как видим линейный алгоритм смог распознать оч малый процент класса отток (меньше 1%), а большой процент правильных ответов(accuracy) получается за счёт большой несбалансированности классов.

## Случайный лес

```python
from sklearn.ensemble import RandomForestClassifier


# n_estimators - количество деревьев
# train_sizes - на какой доле выборки обучаем модель
# max_depth - максимальная глубина
# cv=5 - кросс-валицация на 5 фолдов
# max_feature - количество признаков отобранных для построения дерева (свой для каждого УЗЛА дерева!)
# по умолчанию - корень из всех признаков
parameters_grid = {
    'max_features': ['sqrt', 'log2'],
    'max_depth' : range(5, 10)
}

classifier = RandomForestClassifier(n_estimators = 80, random_state=0, n_jobs=3)

# Будем использовать метод стратификации который делит соотношение классов в обучающей выборке на равное количество
skf = model_selection.StratifiedKFold(n_splits = 5, shuffle = True, random_state = 0)

grid_rfc = model_selection.GridSearchCV(classifier, parameters_grid, scoring = 'roc_auc', cv = skf)
```

```python
%%time
grid_rfc.fit(train_data, train_labels)
```

```python
grid_rfc.best_estimator_
```

```python
print grid_rfc.best_score_
print grid_rfc.best_params_
```

```python
clf2 = RandomForestClassifier(n_estimators = 80, random_state=0, max_depth=9, n_jobs=3)
clf2.fit(train_data, train_labels)
actual_labels2 = clf2.predict(train_data)
```

```python
print "accuracy =", clf2.score(train_data, train_labels)
print "precision =", metrics.precision_score(train_labels, actual_labels2)
print "recall =", metrics.recall_score(train_labels, actual_labels2)
```

```python
(actual_labels2 == 1).sum()
```

## Градиентный бустинг деревьев

```python
from sklearn.ensemble import GradientBoostingClassifier

# learning_rate - контролирует, насколько сильно каждое дерево будет пытаться исправить ошибки предыдущих деревьев.
parameters_grid = {
#     'n_estimators' : range(60, 120, 20),
    'learning_rate' : [0.1, 0.2],
    'max_depth': range(3, 5, 1)
}

classifier = GradientBoostingClassifier(n_estimators=80, random_state=0)

grid_rfc = model_selection.GridSearchCV(classifier, parameters_grid, scoring = 'roc_auc', cv = 5)
```

```python
%%time
grid_rfc.fit(train_data, train_labels)
```

```python
grid_rfc.best_estimator_
```

```python
print grid_rfc.best_score_
print grid_rfc.best_params_
```

```python
clf3 = GradientBoostingClassifier(n_estimators = 80, random_state=0, learning_rate=0.1, max_depth=3)
clf3.fit(train_data, train_labels)
actual_labels3 = clf3.predict(train_data)
```

```python
print "accuracy =", clf3.score(train_data, train_labels)
print "precision =", metrics.precision_score(train_labels, actual_labels3)
print "recall =", metrics.recall_score(train_labels, actual_labels3)
```

```python
def plot_feature_importances_cancer(model):
    n_features = train_data.shape[1]
    plt.figure(figsize(12,55))
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), train_data.columns)
    plt.xlabel(u"Важность признака")
    plt.ylabel(u"Признак")
```

```python
(actual_labels3 == 1).sum()
```

```python
# посмотрим признаки которые вносят самый значимый вклад в модель
plot_feature_importances_cancer(clf3)
```

**NOTE:** Самый большой показатель *ROC AUC* удалось получить с помощью *Градиентного бустинга деревьев* (0.735). Но на всех используемых методах получается довольно низкий показатель полноты (recall).
