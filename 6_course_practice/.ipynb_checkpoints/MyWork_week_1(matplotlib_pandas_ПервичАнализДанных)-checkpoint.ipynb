```python
data = pd.read_csv('orange_small_churn_data.train', sep = ',', header = 0)
labels = pd.read_csv('orange_small_churn_labels.train', sep = ',', header = None, names = ['label'])
print data.shape, labels.shape
data.head()
```

3\. Рассчитайте корреляции переменных с целевой функцией и проанализируйте полученные данные.

Обратите внимание на то, что целевая функция представлляет собой бинарную переменную. Теоретически, это не помешает рассчиать, например, корреляцию Пирсона между непрерывной переменной и целевой функцией, онднако анализиоовать полученные результаты будет сложно, корреляция Пирсона просто не рассчитана на такое применение. Кто уже забыл, как действовать в такой ситуации - вспоминить можно вот здесь: https://www.coursera.org/teach/stats-for-data-analysis/content/edit/lecture/prX3S/video-subtitles


**NOTE:** На самом деле для пар признаков, один из которых непрерывный, а другой — категориальный, вообще не
нужно считать никакой коэффициент корреляции. X1 ∈ R и X2 ∈ {0, 1} будут положительно коррелированы,
если E (X1 | X2 = 1) > E (X1 | X2 = 0).

Таким образом, мерой силы взаимосвязи между X1 и X2 может служить просто разность этих математических ожиданий.

```python
# отделяем числовые и категориальные признаки и удаляем полностью пустые признаки
data_numb = data.iloc[:,0:190].dropna(axis=1, how='all')
data_categ = data.iloc[:,190:].dropna(axis=1, how='all')
```

```python
# нормируем числовые признаки
data_neg_mean = (data_numb.loc[labels['label']==1]/data_numb.loc[labels['label']==1].std()).mean()
data_pos_mean = (data_numb.loc[labels['label']==-1]/data_numb.loc[labels['label']==-1].std()).mean()
```

```python
# получаем корреляцию признаков с labels
corr_outflow = data_pos_mean - data_neg_mean
corr_outflow = corr_outflow.fillna(0)
```

4\. Визуализируйте данные: рассмотрите топ 20 числовых переменных, наиболее сильно коррелирующих с целевой функцией. Для этих переменных постройте:
* Распределения в разрезе классов
* Отобразите объекты в координатах пар признаков

```python
# сортируем по модулю и вытаскиваем 20 лучших корреляций
top20 = corr_outflow.abs().sort_values(ascending = False)[:20]
corr_top20 = corr_outflow[top20.index]
```

```python
corr_top20
```

```python
# Распределения в разрезе классов
fig, axes = plt.subplots(5, 4, figsize=(20,20))

for i, var in enumerate(corr_top20.index):
    var_neg = data_numb[var].loc[labels['label']==1]
    var_pos = data_numb[var].loc[labels['label']==-1]
               
    seaborn.kdeplot(var_neg, color = 'r', ax = axes[int(i / 4) , i % 4], label=u'ушли')
    seaborn.kdeplot(var_pos, color = 'b', ax = axes[int(i / 4) , i % 4], label=u'остались')
    axes[int(i / 4), i % 4].set(xlabel=var)
    
fig.tight_layout()
```

```python
# объекты в координатах пар признаков
# NOTE: т.к. графики будут нечитаемы, то сравниваем лучшие 7 признаков
merged_df = pd.concat([data_numb[corr_top20.index[:7]], labels], axis = 1, sort=False)
seaborn.pairplot(merged_df, hue="label", diag_kind="kde")
```

```python
# удалим те колонки где больше 10 категорий и меньше 2 (эти колонки не информативны)
name_del = [name for name, var in data_categ.iteritems() if var.value_counts(dropna=True).shape[0] > 11 or var.value_counts(dropna=True).shape[0] < 2]
new_categ = data_categ.drop(labels=name_del, axis=1)

fig, axes = plt.subplots(new_categ.shape[1], 2, figsize=(20, new_categ.shape[1] * 4))

i=0
for name, var in new_categ.iteritems():

    var_neg = var[labels['label']==1].dropna()
    var_pos = var[labels['label']==-1].dropna()
    
#     print var_neg.value_counts()
#     print var_pos.value_counts()
#     break
    
    var_neg.value_counts(normalize=True, dropna=True).plot('bar', ax=axes[i,0])
    var_pos.value_counts(normalize=True, dropna=True).plot('bar', ax=axes[i,1])
    axes[i,0].set_title(u'%s, ушли' % name)
    axes[i,1].set_title(u'%s, не ушли' % name)
    i+=1
    
fig.tight_layout()
```
