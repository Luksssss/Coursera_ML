## Инструкции

1\. Загрузите данные `orange_small_churn_data.train` и `orange_small_churn_labels.train`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from scipy import stats
```

```python
labels=pd.read_csv('orange_small_churn_labels.txt', header=None)
data=pd.read_csv('orange_small_churn_data.txt')
data.head()
```

2\. Рассчитайте доли классов `отток` и `не отток`.

```python
print('отток:',labels[0].value_counts(normalize=True)[-1])
print('не отток:',labels[0].value_counts(normalize=True)[1])
```

3\. Рассчитайте корреляции переменных с целевой функцией и проанализируйте полученные данные.

Обратите внимание на то, что целевая функция предсталвляет собой бинарную переменную. Теоретически, это не помешает рассчиать, например, корреляцию Пирсона между непрерывной переменной и целевой функцией, онднако анализиоовать полученные результаты будет сложно, корреляция Пирсона просто не рассчитана на такое применение. Кто уже забыл, как действовать в такой ситуации - вспоминить можно вот здесь: https://www.coursera.org/teach/stats-for-data-analysis/content/edit/lecture/prX3S/video-subtitles 

```python
feat_num=list(data)[:190] ## numerical features
feat_cat=list(data)[190:] ## categorial features
class_n1_idx = labels[labels[0]==-1].index ## class -1 data
class_p1_idx = labels[labels[0]==1].index ## class +1 data
```

```python
data_num=data[feat_num].dropna(axis=1, how='all') # drop columns with all NA
data_num = data_num.fillna(data_num.mean())
```

```python
data_n1_mean = ((data_num.loc[class_n1_idx])/data_num.loc[class_n1_idx].std()).mean()  # normalize data for class -1 
data_p1_mean = ((data_num.loc[class_p1_idx])/data_num.loc[class_n1_idx].std()).mean()  # normalize data for class +1 
```

```python
diff_statdf=pd.DataFrame(abs(data_n1_mean-data_p1_mean)) # to dataframe
diff_statdf.sort_values(0,inplace=True, ascending=False)
diff_statdf.head(20)
```

```python
data_num['response'] = labels[0]
data_cat['response'] = labels[0]
```

```python
top20_features = list(diff_statdf.iloc[:20].index)
bottom10_features = list(diff_statdf.iloc[-10:].index)
random10_features = list(diff_statdf.iloc[np.random.randint(0,len(diff_statdf),10)].index)
```

4\. Визуализируйте данные: рассмотрите топ 20 числовых переменных, наиболее сильно коррелирующих с целевой функцией. Для этих переменных постройте:
* Распределения в разрезе классов
* Отобразите объекты в координатах пар признаков

```python
rand_idx = np.random.randint(0,40000,1000)
seaborn.pairplot(data_num[top20_features[:5]+['response']].iloc[rand_idx], hue='response')  # pairplot for rand 1000 elements
plt.draw()
```

```python
fig, axes = plt.subplots(5, 4, figsize=(20,20))

for i,elm in enumerate(top20_features):
    data_n1 = (data_num[elm].iloc[class_n1_idx])
    data_p1 = (data_num[elm].iloc[class_p1_idx])

    seaborn.kdeplot(data_n1, color = 'r', ax = axes[int(i / 4) , i % 4], label='no churn')
    seaborn.kdeplot(data_p1, color = 'b', ax = axes[int(i / 4) , i % 4], label='churn')
    axes[int(i / 4), i % 4].set(xlabel=elm)
fig.tight_layout()
```

7\. Посмотрите на категориальные переменные: постройте и сравните гистограммы значений категориальных переменных для разных классов.

```python
important_features = data[feat_cat].isna().sum()/40000  < 0.5
important_features = important_features[important_features.values == True].index
important_data = data[important_features]
important_data.head()
```

```python
cat_counts = pd.Series([len(data[important_features[i]].value_counts()) for i in range(len(important_features))], index = important_features)
# для отображения возмем признаки у которых не более 10 уникальных значений
features = list(cat_counts[cat_counts<=10].index)
```

#### 

```python
fig, axes = plt.subplots(int(len(important_features)/2), 2, figsize=(20,len(important_features) * 4))
for i, elm in enumerate(features):
    data[elm] = data[elm].dropna().apply(lambda x: str(x))
    data[elm].iloc[class_n1_idx].value_counts(normalize=True,dropna=True).plot('bar', ax=axes[i,0])
    data[elm].iloc[class_p1_idx].value_counts(normalize=True,dropna=True).plot('bar', ax=axes[i,1])
    axes[i,0].set_title('%s, no churn' % elm)
    axes[i,1].set_title('%s, churn' % elm)
fig.tight_layout()
```
