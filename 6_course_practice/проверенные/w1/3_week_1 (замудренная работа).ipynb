3\. Рассчитайте корреляции переменных с целевой функцией и проанализируйте полученные данные.

Обратите внимание на то, что целевая функция предсталвляет собой бинарную переменную. Теоретически, это не помешает рассчиать, например, корреляцию Пирсона между непрерывной переменной и целевой функцией, онднако анализиоовать полученные результаты будет сложно, корреляция Пирсона просто не рассчитана на такое применение. Кто уже забыл, как действовать в такой ситуации - вспоминить можно вот здесь: https://www.coursera.org/teach/stats-for-data-analysis/content/edit/lecture/prX3S/video-subtitles 

#### Предварительные рассуждения
Рассуждать будем так, у нас имеется 230 переменных, из которых первые 190 переменных - числовые, и оставшиеся 40 переменные - категориальные. Сам отклик - бинарный. Корреляции для пар - вещественные-категориальные считать нельзя, они плохо интерпретируются, вместо этого расчитывается разность математических ожиданий:

<center>$K_{abs} = E (X | Y = 1) − E (X | Y = -1)$</center>

Для того, чтобы разности как-то сравнивать, приведем значение вещественных признаков к шкале [0,1] (нормируем) по формуле:

<center>$X_{norm} = \frac{X-X_{min}}{X_{max}-X_{min}}$</center>

Легко видеть, что для разности средних это означает:

<center>$K_{norm} = \frac{E(X | Y = 1) − E(X_1 | Y = -1)}{X_{max}-X_{min}}$ (1)</center>

В случае 40 оставшихся категориальных признаков расчитаем коэффициент V Крамера, составив соответствующие таблицы сопряженностей, для вычисления используем формулу:

<center>$\Phi_c (X_1^n , X_2^n) = \sqrt\frac{χ^2(X_1^n , X_2^n)}{n(min(K_1 , K_2)−1)} $ (2)</center>


```python
data['label'] = labels.label
```

```python
# Получим DataFrame для вещественых признаков
df_num  = data.iloc[:,[i for i in range(190)]+[-1]]
# Абсолютные разности
K_abs = df_num[df_num.label == 1].iloc[:,:190].mean() - df_num[df_num.label == -1].iloc[:,:190].mean()
# Нормированные разности
K_norm = K_abs/(df_num.iloc[:,:190].max() - df_num.iloc[:,:190].min())
```

```python
print('Значения разницы по абсолютной шкале:')
K_abs = pd.DataFrame(K_abs, columns=['k']).dropna()
#Добавим колонку с абсолютным значением разности и отсортируем по ней
K_abs['abs_k'] = K_abs.k.apply(abs)
K_abs.sort_values(by ='abs_k' , ascending=False)
```

Становсится очевидным, что подобная сотртировка по абсолютным значением мало о чем нам говорит, поэтому посмотрим на значениея в относительной шкале

```python
print('Значения разницы по относительной шкале:')
K_norm = pd.DataFrame(K_norm, columns=['k']).dropna()
#Добавим колонку с абсолютным значением разности и отсортируем по ней
K_norm['abs_k'] = K_norm.k.apply(abs)
K_norm.sort_values(by ='abs_k' , ascending=False)
```

Посмотрим на статистическую значимость разницы средних для каждой переменной, для этого проверим нулевую гипотезу о том, что средние совпадают против двухсторонней альтернативы:

$H_0: mean(X|label=1) = mean(X|label=-1)$

$H_1: mean(X|label=1) \neq mean(X|label=-1)$

Проверять будем только для тех признаков, которые на своих срезах label пройдут тест на критерий Шапиро-Уилка:

 $H_0\colon$ данные распредлены нормально
 
 $H_1\colon$ данные распредлены не нормально.

```python
from statsmodels.stats.weightstats import *
import warnings
warnings.filterwarnings('ignore')
```

```python
shapiro = []
ttests = []
for col in K_norm.index.values:
    shapiro.append(
          (stats.shapiro(data[(data.label==-1)&(data[col].notna())][[col]])[1] < 0.05)&
          (stats.shapiro(data[(data.label==1)&(data[col].notna())][[col]])[1] < 0.05)         
         )
    
    ttests.append(stats.ttest_ind(data[(data.label==-1)&(data[col].notna())][[col]],
                                 data[(data.label==1)&(data[col].notna())][[col]],
                                 equal_var = False
                                )[1][0])

K_norm['shapiro'] = [1 if i else 0 for i in shapiro]
K_norm['p_value'] = ttests
K_norm['correlated'] = [1 if i < 0.05 else 0 for i in ttests]

print('Сотртировка результата в порядке прохождения тестов и уменьшения разницы')

K_norm.sort_values(by =['shapiro','correlated','abs_k'] , ascending=[False,False,False])
```

```python
print('Значимая зависимость установлена для переменных, распределенных нормально:')
K_norm[(K_norm.shapiro == 1)&(K_norm.correlated == 1)].index.values
```

```python
#Получим DataFrame для категориальных признаков
df_cat = data.iloc[:,190:]
#удалим полностью незаполненые колонки
cnts_cat = pd.DataFrame(df_cat.count(), columns=['cnt'])
print("Колонки, которые полностью не заполнены, исключены из набора по расчету корреляции:")
cnts_cat = cnts_cat[cnts_cat.cnt==0].index.values
print(cnts_cat)
df_cat.drop(cnts_cat, axis=1, inplace=True)
```

```python
# Составление матриц смежности
matrixes = {}
for col in df_cat.iloc[:,:-1]: 
    matrixes[col]  = df_cat.loc[:, [col,'label']]
    matrixes[col]['cnt'] = 1
    #Чтобы не удалять строки не срдержащих значение, представим их как отдельный признак NODATA
    matrixes[col].fillna('NODATA', inplace=True)
    matrixes[col] = matrixes[col].pivot_table(index='label', columns=col, 
                                              values='cnt', dropna=False, aggfunc='sum').fillna(0) 
    print("Форма матрицы сопряженноти: label-"+col+' ', 
          matrixes[col].shape,
          '\tmin(label,'+col+')=',
          min(matrixes[col].shape[0], matrixes[col].shape[1]),
          '\tколичество элементов n=',int(matrixes[col].sum().sum())
         ) 
```

```python
#Выведим последнюю мастрицу смежности
matrixes[col]
```

Из формулы (2) и по выведеным данным следует:
* $min(K_1 , K_2)−1 = 1$
* $n = 40000$
* $\Phi_c (X_1^n , X_2^n) = \sqrt\frac{χ^2(X_1^n , X_2^n)}{40000}$
* Для расчета $χ^2$ воспользуемся методом chi2_contingency из библиотеки scipy.stats
* Проверим также нулевую гипотезу о корреляции каждого признака в отдельности 

$H_0$ - не коррелирует, $H_1$ - коррелирует

```python
import scipy.stats as stats
import math
```

```python
K_cat = {}
for k in matrixes.keys():
    
    arr = stats.chi2_contingency(matrixes[k])
    K_cat[k] = {'k': math.sqrt(arr[0]/40000), 'p_value':arr[1]}
    
K_cat = pd.DataFrame(K_cat).T
K_cat['correlated'] = K_cat.p_value.apply(lambda x: 1 if x < 0.05 else 0)

#K_cat.columns=['k'], 
print("Отсортированные кореляции по Крамеру для категориальных признаков\nи целевой переменной в порядке отвержения нулевой гипотезы\nи убывания коэф. Крамера")
K_cat.sort_values(by =['correlated', 'k'] , ascending=[False,False])
```

```python
print('Значимая корреляция наблюдаемая для категориальных признаков с целевой переменной:')
K_cat[K_cat.correlated == 1].index.values
```

```python
# функция для отображения распредления вещественных признаков на целевую переменную
def num_plot_with_label(features, cols=4):
    rows = len(features)//cols
    if(len(features)%cols > 0):
        rows += 1 
    width = 16
    sq_a = width/cols
    
    plt.figure(figsize(width, rows*sq_a))
    
    grid = plt.GridSpec(rows, cols, hspace=0.5, wspace=0.3)
    for row in range(rows):
        for col in range(cols):
            indx = row*cols+col
            if(indx > (len(features)-1)):
                break
            
            pylab.subplot(grid[row,col])
            df = data[[features[indx],'label']].dropna()
            df[features[indx]][df.label==1].plot.kde(color='g', label=u'Не отток')
            df[features[indx]][df.label==-1].plot.kde(color='r', label=u'Отток')
            plt.legend()
            plt.grid()
            plt.xlabel(features[indx])
            plt.ylabel(u'Доля')
            plt.title(u'Зависимость label от '+features[indx])
```

На диаграммах хорошо прослеживается сдвиг среднего для разных классов целевой переменной, для более наглядной картины построим boxplot диаграммы для каждого признака в разрезе классов

```python
# функция для отображения распредления вещественных признаков на целевую переменную (boxplot)
def num_boxplot_with_label(features, cols=4):
    rows = len(features)//cols
    if(len(features)%cols > 0):
        rows += 1 
    width = 16
    sq_a = width/cols
    
    plt.figure(figsize(width, rows*sq_a*2))
    labels = (u'Не отток\n(label=-1)',
         u'Отток\n(label=1)')
    
    grid = plt.GridSpec(rows, cols, hspace=0.2, wspace=0.3)
    for row in range(rows):
        for col in range(cols):
            indx = row*cols+col
            if(indx > (len(features)-1)):
                break
            
            pylab.subplot(grid[row,col])
            df = data[[features[indx],'label']].dropna()
            pylab.boxplot([
              df[df.label ==-1][features[indx]].values, 
              df[df.label == 1][features[indx]].values
              ],vert=True, labels=labels)
            
            plt.ylabel(features[indx])
            plt.xlabel('label')
            plt.title(u'Зависимость label от '+features[indx])
```

```python
#Воспользуемся ранее построенными матрицами смежности matrixes, отнормируем с помощью полученных делитлей
norm_matrixes = {}
for indx in matrixes:
    tmp = matrixes[indx].copy()
    tmp[tmp.index.values==-1] = tmp[tmp.index.values==-1]/nocrn_scale
    tmp[tmp.index.values==1] = tmp[tmp.index.values==1]/crn_scale
    norm_matrixes[indx] = tmp
print("Последний обработанный элемент:")
norm_matrixes[indx]
```

#### * Как вы думаете, окажется ли отбор признаков полезным для построения модели?
Вопрос философский. Если строится модель с использованием линейных классификаторов - очевидно что отбор признаков необходим для избегания "кошмара размерностей" и удаления повторяющихся признаков, которые сильно между собой коррелируют.
